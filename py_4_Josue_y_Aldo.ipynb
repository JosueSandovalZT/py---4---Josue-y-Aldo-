```python
import zipfile
import pandas as pd
import io
from google.colab import files
import os

# 1. Subir archivo ZIP
uploaded = files.upload()

# Obtener el nombre del archivo subido
zip_filename = list(uploaded.keys())[0]
print(f"Archivo subido: {zip_filename}")

# 2. Descomprimir y explorar contenido
with zipfile.ZipFile(zip_filename, 'r') as zip_ref:
    # Mostrar contenido del ZIP
    file_list = zip_ref.namelist()
    print("\n Contenido del archivo ZIP:")
    for file in file_list:
        print(f"  - {file}")

    # Buscar archivos CSV
    csv_files = [f for f in file_list if f.lower().endswith('.csv')]

    if csv_files:
        print(f"\n Se encontraron {len(csv_files)} archivos CSV:")

        # Cargar el primer archivo CSV en un DataFrame
        first_csv = csv_files[0]
        print(f" Cargando: {first_csv}")

        with zip_ref.open(first_csv) as csv_file:
            # Leer el archivo CSV directamente en un DataFrame
            df = pd.read_csv(csv_file)

        print("DataFrame creado exitosamente!")

    else:
        print("No se encontraron archivos CSV en el ZIP")
        df = None

# 3. Mostrar información del DataFrame
if df is not None:
    print("\n" + "="*50)
    print("INFORMACIÓN DEL DATAFRAME")
    print("="*50)

    print(f"Dimensiones: {df.shape[0]} filas × {df.shape[1]} columnas")
    print(f"\n Primeras 5 filas:")
    print(df.head())

    print(f"\n Información general:")
    print(df.info())

    print(f"\n Estadísticas descriptivas:")
    print(df.describe())

    print(f"\n Valores nulos por columna:")
    print(df.isnull().sum())
```
